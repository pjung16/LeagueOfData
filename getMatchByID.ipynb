```python
challenger_players = "https://na1.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5?api_key=RGAPI-96bceb3e-271d-4f81-b3c8-03413e8bb757"
```

```python
import urllib.request
import json
with urllib.request.urlopen(challenger_players) as url:
    data = json.loads(url.read().decode())
data
```

```python
players = data['entries']
players
```

```python
playerIDs = []
for player in players:
    playerIDs.append(player['summonerId'])
playerIDs
```

```python
len(playerIDs)
```

```python
api_link = "https://na1.api.riotgames.com/lol/summoner/v4/summoners/temp_name?api_key=RGAPI-96bceb3e-271d-4f81-b3c8-03413e8bb757"


```

```python
import time
encryptedIDs = []
for i in playerIDs:
    print(api_link.replace('temp_name', i))
    with urllib.request.urlopen(api_link.replace('temp_name', i)) as url:
        encryptedIDs.append(json.loads(url.read().decode()))
        time.sleep(1.2)
encryptedIDs
```

```python
accountIDs = []
for i in encryptedIDs:
    accountIDs.append(i['accountId'])
accountIDs
```

```python
api_link = "https://na1.api.riotgames.com/lol/match/v4/matchlists/by-account/eID?api_key=RGAPI-96bceb3e-271d-4f81-b3c8-03413e8bb757"
```

```python
with urllib.request.urlopen(api_link.replace('eID', accountIDs[0])) as url:
        matches = json.loads(url.read().decode())
matches
```
